```python
df_ca = pd.read_csv('airbnbSeattle/calendar.csv')

df_ca['date'] = pd.to_datetime(df_ca['date'])
df_ca['available'] = df_ca['available'].replace({'t': True, 'f': False})
df_ca['price'] = df_ca['price'].str.replace(',', '').str.replace('$', '').astype(float)

print(df_ca.shape)
df_ca.head()
```

```python
df = df_ca.merge(df_li.drop(columns='price'), left_on='listing_id', right_on='id')

print(df.shape)
df.head()
```

```python
df_group = df.groupby(['neighbourhood', pd.Grouper(key='date', freq='M')])\
.agg({'price': 'mean', 'available': 'mean'}).reset_index()

df_group.head()
```

```python
most_common = df['neighbourhood'].value_counts().head().index.values

f, axs = plt.subplots(2, 1, figsize=(12, 10))
for mc in most_common:
    query = df_group[(df_group['neighbourhood'] == mc) & (df_group['date'].dt.year == 2016)]
    query.plot(x='date', y='price', ax=axs[0], label=mc)
    query.plot(x='date', y='available', ax=axs[1], label=mc)
#     ax.plot(query['date'], query[('price', 'count')], label=mc)

axs[0].set_ylabel('Mean price')
axs[1].set_ylabel('Availability');
```
